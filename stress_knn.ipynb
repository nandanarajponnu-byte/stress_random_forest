```python
import pandas as pd
data=pd.read_csv(r"C:\Users\nanda\Downloads\Stress-Lysis.csv")
print(data)
data.isnull().sum()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
model2=KNeighborsClassifier(n_neighbors=5)
x = data[['Humidity','Temperature','Step count']]
y = data['Stress Level']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
model2.fit(x_train, y_train)
y_pred2= model2.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, y_pred2))
```

```python
a=int(input("Enter humidity level"))
b=int(input("Enter temperature level"))
c=int(input("Enter number of steps"))
model2.predict([[a,b,c]])
```
